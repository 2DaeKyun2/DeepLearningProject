```python
from ultralytics import YOLO
# 1. YOLOv8m Train9번 > 23번 yolov8m 에폭30, 배치 6, 이미지사이즈=default
# 2. YOLOv8m train23번 > 27번 yolov8m 에폭 60 배치5
# 3. YOLOv8m train27번 > 29번 yolov8m 에폭 10 배치5
# 4. YOLOv8m train 29번 > 30번 yolov8m 에폭 100, 배치5


##### YOLOv5s train 32번 > 33번 누적에폭80+30 배치 5, imgsize=940  #####
##### YOLOv5s train 33번 > 34번 누적에폭 110+40 배치5 imgsize=940   #####
model = YOLO('runs/detect/train33/weights/best.pt')

# 학습 - train()메소드 호출
model.train(data='Bird_Drone_Detection.v2i.yolov5(220320)/data.yaml', epochs=40, patience=10,  # 지정한 epoch만큼 성능이 개선이 안되면 조기종료.
            batch=5,    # batch size
            imgsz=940,    # input image size(defalult(640))
            optimizer='Adam')
```
